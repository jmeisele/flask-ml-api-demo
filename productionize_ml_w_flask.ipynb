```python
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import pickle
import requests
import json
import pandas as pd
from requests.auth import HTTPProxyAuth
import os
```

```python
#Import the iris dataset and split between variables and target variable
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
print(iris['target_names'])
```

```python
print(iris.data)
```

```python
print(iris.target)
```

```python
print(iris.DESCR)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
rfc = RandomForestClassifier(n_estimators = 100, n_jobs = 2)
```

```python
rfc.fit(X_train, y_train)
```

```python
print("Accuracy = %0.2f" % accuracy_score (y_test, rfc.predict(X_test)))
print(classification_report(y_test, rfc.predict(X_test)))
```

## Model Serialization / Marshalling

```python
pickle.dump(rfc, open("iris_rfc.pkl", "wb"))
```

```python
my_random_forest = pickle.load(open("iris_rfc.pkl", 'rb'))
```

```python
my_random_forest
```

```python
print(classification_report(y_test, my_random_forest.predict(X_test)))
```
